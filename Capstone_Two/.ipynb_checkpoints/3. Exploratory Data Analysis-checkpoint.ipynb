# Exploratory Data Analysis

This notebook is used to begin exploring the tweet data a little bit deeper. The optimal times of day to tweet were discovered based on the average number of favorites a tweet tweeted at any hour of the day receives.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime
```

```python
tweet_data = pd.read_csv('clean_tweet_data.csv',index_col=0,parse_dates=['created_at'])
```

```python
tweet_data.head()
```

First let's sum the total amount of favorites and retweets received each day of the week. The count on the text column is used to determine the number of tweets for that day, which will subsequently be used to determine favorites/tweet and retweets/tweet.

```python
df1 = tweet_data[['favorites','retweets','Day_of_Week']].groupby('Day_of_Week').sum()
```

```python
df2 = tweet_data[['text','Day_of_Week']].groupby('Day_of_Week').count()
```

```python
data_per_day = df1.join(df2, on='Day_of_Week')
data_per_day
```

```python
data_per_day['Favorites/Tweet'] = round(data_per_day.favorites/data_per_day.text,2)
data_per_day['Retweets/Tweet'] = round(data_per_day.retweets/data_per_day.text,2)
data_per_day.rename(columns={'favorites':'Favorites','retweets':'Retweets','text':'# of Tweets'})
```

Let's plot the daily distribution to see if there are any significantly better days to tweet.

```python
ax = data_per_day.plot(kind='bar',y=['Favorites/Tweet','Retweets/Tweet'],color=['navy','lightcoral'],figsize=(10,7))
sns.despine()
ax.set_title('Favorites/Tweet and Retweets/Tweet for Days of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('# of Retweets/Favorites')
```

It appears that thursday (3) and friday (4) are on average the worst days and sunday (6) is the best day of the week for tweets.

I need to make sure there is a relatively even distribution of tweets per day for each artist so that one artist's larger follower count doesn't skew the results.

```python
twts_per_day_per_artist = tweet_data.pivot_table(index = 'screen_name',columns = 'Day_of_Week',values='text',aggfunc='count')
twts_per_day_per_artist
```

```python
twts_per_day_per_artist.div(twts_per_day_per_artist.sum(axis=1), axis=0)
```

It seems like the daily tweet distribution is quite balanced and should not skew the results significantly.

Now let us look at tweet interaction on an hourly basis

```python
tweet_data['time_of_day'] = tweet_data['created_at'].dt.hour
tweet_data.head()
```

The same procedure for obtaining the favorites/tweet and retweets/tweet on a daily basis as above is used to determine the results on an hourly basis.

```python
df1 = tweet_data[['favorites','retweets','time_of_day']].groupby('time_of_day').sum()
df2 = tweet_data[['text','time_of_day']].groupby('time_of_day').count()
data_per_hour = df1.join(df2, on='time_of_day')
data_per_hour
```

```python
data_per_hour['Favorites/Tweet'] = round(data_per_hour.favorites/data_per_hour.text,2)
data_per_hour['Retweets/Tweet'] = round(data_per_hour.retweets/data_per_hour.text,2)
```

Plotting the results:

```python
ax = data_per_hour.plot(kind='bar',y=['Favorites/Tweet','Retweets/Tweet'],color=['navy','lightcoral'],figsize=(10,7))
plt.legend(loc='center',bbox_to_anchor=(0.5,0.8))
sns.despine()
ax.set_title('Favorites/Tweet and Retweets/Tweet for Hours of the Day')
plt.xlabel('Hour of the Day')
plt.ylabel('# of Retweets/Favorites')
```

As we can see in the plot above, there is a clear dip in tweet performance during the day from around 8 am - 5 pm, which is easily explained by the fact that those cover the normal working hours for most people. The optimal time to tweet seems to be from 2 am - 7 am EST.
