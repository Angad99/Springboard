# Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import datetime
```

```python
tweet_data = pd.read_csv('clean_tweet_data.csv',index_col=0,parse_dates=['created_at'])
```

```python
tweet_data.head()
```

```python
df1 = tweet_data[['favorites','retweets','Day_of_Week']].groupby('Day_of_Week').sum()
```

```python
df2 = tweet_data[['text','Day_of_Week']].groupby('Day_of_Week').count()
```

```python
data_per_day = df1.join(df2, on='Day_of_Week')
data_per_day
```

```python
data_per_day['faves/tweet'] = data_per_day.favorites/data_per_day.text
data_per_day['rts/tweet'] = data_per_day.retweets/data_per_day.text
data_per_day
```

```python
data_per_day.plot(y=['faves/tweet','rts/tweet'])
```

```python
twts_per_day_per_artist = tweet_data.pivot_table(index = 'screen_name',columns = 'Day_of_Week',values='text',aggfunc='count')
twts_per_day_per_artist
```

```python
twts_per_day_per_artist.div(twts_per_day_per_artist.sum(axis=1), axis=0)
```

```python
tweet_data['time_of_day'] = tweet_data['created_at'].dt.hour
tweet_data.head()
```

```python
df1 = tweet_data[['favorites','retweets','time_of_day']].groupby('time_of_day').sum()
df2 = tweet_data[['text','time_of_day']].groupby('time_of_day').count()
data_per_hour = df1.join(df2, on='time_of_day')
data_per_hour
```

```python
data_per_hour['faves/tweet'] = data_per_hour.favorites/data_per_hour.text
data_per_hour['rts/tweet'] = data_per_hour.retweets/data_per_hour.tex
```

```python
data_per_hour.plot(y=['faves/tweet','rts/tweet'])
```

```python
data_per_hour.sort_values(by='faves/tweet',ascending = False)
```
