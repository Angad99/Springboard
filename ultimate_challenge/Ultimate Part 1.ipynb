# Ultimate Data Challenge - Part 1

Import the necessary packages:

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Read in the data and inspect it:

```python
data = pd.read_json('logins.json')
```

```python
data.head()
```

```python
data.describe()
```

Add a "tally" column for each summing of # of logins per time interval.

```python
data['tally'] = 1
```

```python
data.head()
```

Resample the data to 15 minute intervals:

```python
data_15min = data.set_index('login_time').resample('15T').count()
```

```python
data_15min.plot(figsize=(12,6))
```

There appears to be some pattern to the data. Let's investigate further.

Plot just two days:

```python
fig, ax = plt.subplots(figsize=(12,6))
data_15min['1970-01-02':'1970-01-02'].plot(ax=ax)
data_15min['1970-01-03':'1970-01-03'].plot(ax=ax)
```

It seems like there are increased logins in the late night (21:00-04:00) and a slightly smaller spike around noon. Let's see if this holds true in general for all the data.

Aggregate data by daily time and visualize using a box plot:

```python
data_15min.index[1].time()
```

```python
data_15min['time'] = [t.time() for t in data_15min.index]
```

```python
data_15min.sort_values(by='time',inplace=True)
fig, ax = plt.subplots(figsize=(15,8))
sns.boxplot(ax=ax,x='time',y='tally',data=data_15min)
plt.xticks(rotation=80)
```

From the above box plot we can see that there is a spike in logins from 10:45-13:00 as well as a night spike beginning around 20:30, continuing until 01:00, whereby it begins to decrease until 05:30. The lowest login period is from 05:30-09:00 and there is very little variability in that time as well.
