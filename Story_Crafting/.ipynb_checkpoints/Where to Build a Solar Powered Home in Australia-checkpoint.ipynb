# Where to Build a Solar Powered Home in Australia

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Dataset obtained from: https://www.kaggle.com/jsphyg/weather-dataset-rattle-package

```python
data = pd.read_csv('weatherAUS.csv',parse_dates = [0])
```

```python
data.head()
```

```python
data.describe()
```

```python
data.groupby(['Location']).mean().head(10)
```

```python
sunshine_cities = data.groupby(['Location']).mean().sort_values(by='Sunshine',ascending=False)
sunshine_cities.drop(['PerthAirport','PearceRAAF'],inplace=True)
```

```python
columns = ['MinTemp','MaxTemp','Rainfall','Sunshine','Humidity9am','Humidity3pm','Cloud9am','Cloud3pm','Temp9am','Temp3pm']
top_cities = sunshine_cities.head(15)[columns]
```

```python
top_cities
```

```python
top_cities.plot.bar(y='Sunshine')
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(y=top_cities.index,x='Sunshine',data=top_cities, palette = 'viridis')
ax.set_title('Average Daily Sunshine in Australian Cities 2007-2017',fontsize=16)
ax.set_xlabel('Average Daily Sunshine (hrs)',fontsize=16)
ax.set_ylabel('City',fontsize=16)
```

```python
top_10 = top_cities.head(10)
```

```python
top_10.index
```

Population data obtained from Google.

```python
top_10['Population'] = (26390,150,7383,1985000,3990,132045,30016,178860,65258,2280000)
```

```python
top_10.sort_values(by='Population',inplace=True)
```

```python
plt.figure(figsize=(10,8))
ax = sns.barplot(y=top_10.index,x='Population',data=top_10, palette = 'viridis')
ax.set_title('Population of Australian Cities in 2016',fontsize=16)
ax.set_xlabel('Population',fontsize=16)
ax.set_ylabel('City',fontsize=16)
```
